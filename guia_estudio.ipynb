##Vista vs copia

```python
vista = A[0, :] # es una vista
copia = A[0, :].copy() # es una copia independiente
print("VISTA\n",vista)
print("COPIA\n",copia,"fin\n")
vista[0] = 999 # modifica A también!
copia[0] = 888 # NO modifica A
print(A[0, :]) # [999 2 3 4] ← vista modificó el original
print(np.shares_memory(A, vista)) # True
print(np.shares_memory(A, copia)) # False
```

##Reshape — Cambio de Forma

```python
import numpy as np
v = np.arange(12)# del 0 al 11 (n-1)
print(v)
```

##Reshape básico

```python
M2x6 = v.reshape(2, 6) # 2 filas, 6 columnas
print(M2x6,"\n")

M3x4 = v.reshape(3, 4) # 3 filas, 4 columnas
print(M3x4,"\n")
M4x3 = v.reshape(4, 3) # 4 filas, 3 columnas
print(M4x3,"\n")
M2x2x3= v.reshape(2, 2, 3) # tensor 3-D
print(M2x2x3,"\n")
```

##Uso de -1 (dimension automatica)

```python
print(v,"\n")
# ?x3: NumPy calcula ? = 12/3 = 4 → shape (4,3)
auto = v.reshape(-1, 3)
print(auto.shape) # (4, 3)
auto2 = v.reshape(2, -1) # 2 x ?: ? = 12/2 = 6 → shape (2, 6)
print(auto2.shape) # (2, 6
```

##flatten vs ravel

```python
print(v,"\n")
M = v.reshape(3, 4)
flat = M.flatten() # siempre COPIA
print("copia",flat)
rav = M.ravel() # VISTA si es posible
flat[0] = 999
print(M[0, 0]) # 0 → flatten no modifica el original
rav[0] = 777
print(M[0, 0]) # 777 → ravel SÍ modifica el original (es vista)
```

##Concatenación y Apilamiento de Arrays

```python
import numpy as np
A = np.array([[1, 2, 3],
[4, 5, 6]])
B = np.array([[7, 8, 9],
[10, 11, 12]])
```

Concatenacón


```python
vertical = np.concatenate([A, B], axis=0) # apilar filas
print(vertical.shape) # (4, 3)
horizontal = np.concatenate([A, B], axis=1) # apilar columnas
print(horizontal.shape) # (2, 6
```

vstack y hstack

```python
v1 = np.array([1, 2, 3])
v2 = np.array([4, 5, 6])
print("v1:", v1)
print("v2:", v2)
#vstack apila hacia abajo
print(np.vstack([v1, v2])) # [[1 2 3], [4 5 6]] → shape (2,3)
#hstack junta a los lados
print(np.hstack([v1, v2])) # [1 2 3 4 5 6] → shape (6,)
```

stack — crea nuevo eje

```python
 # shape (2, 2, 3) — nuevo eje al frente
capas = np.stack([A, B], axis=0)
print(capas.shape)
```

argsort

```python
v = np.array([5, 2, 8, 1, 9, 3])
print("v: ",v)
idx = np.argsort(v) # [3 1 5 0 2 4] — índices que ordenarían v
print(v[idx]) # [1 2 3 5 8 9]
```

Reordenar dos arrays en paralelo

```python
import numpy as np
nombres = np.array(['Carlos','Ana','Diana','Pedro','Eva','Luis'])
notas = np.array([75, 92, 88, 61, 95, 78])

# obtener índices que ordenan las notas de mayor a menor
orden = np.argsort(notas)[::-1]

print("nombres ordenados por nota",nombres[orden])

print("notas ordenadas",notas[orden])
```
