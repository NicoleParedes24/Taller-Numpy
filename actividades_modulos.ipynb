###Ejercicio 3: Define una función 'resumen(lista)' que retorne un diccionario con: longitud, suma, promedio, mínimo y máximo de la lista.


```python

def resumen(lista):
    return {
        'longitud': len(lista),
        'suma ': sum(lista),
        'promedio': sum(lista) / len(lista) if lista else 0,
        'minimo': min(lista) if lista else None,
        'maximo': max(lista) if lista else None
    }
```

```python

print(resumen([1, 2, 3, 4, 3]))
```

###Ejercicio 4: Demuestra la diferencia vista vs. copia: modifica un elemento en un slice y verifica que el array original también cambia.


```python
v = np.arange(12)

M = v.reshape(3, 4)
#es la copia -flatten
flat = M.flatten()
#es la vista
rav = M.ravel()
flat[0] = 999
# flatten no modifica el original->0
print(M[0, 0])
rav[0] = 777
# 777 -> ravel(vista) si lo modifica
print(M[0, 0])

```


###Ejercicio 4: Crea dos matrices 3×4 y concaténalas vertical y horizontalmente.

```python
m1=np.random.randint(1,9,(3,4))
m2=np.random.randint(9,15,(3,4))
print("matriz 1: \n",m1)

print("matriz 2\n",m2)

#por filas axis=0
concatenar1=np.concatenate([m1,m2],axis=0)
print("Concatenacion verticalmente: \n", concatenar1)

#por columnas axis=1
concatenar2=np.concatenate([m1,m2],axis=1)
print("Concatencacion horizontalmente: \n",concatenar2)
```

Luego divide el resultado vertical en 3 partes iguales con vsplit


```python
print(concatenar1)
#divide el concatenar1 en 3
div=np.vsplit(concatenar1,3)
print(div)
```
